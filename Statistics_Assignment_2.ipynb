1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.


```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data
sample_mean = 68
sample_std = 5
sample_size = 30
population_mean = 70 # Known population mean

# Calculate the Z-statistic and P-value
# ztest(x1, x2, value) where x1 is the sample, x2 is None for one-sample z-test, value is the population mean
z_statistic, p_value = ztest(x1=np.random.normal(sample_mean, sample_std, sample_size), value=population_mean)

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("There is significant evidence to conclude that the sample mean is different from the population mean.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("There is no significant evidence to conclude that the sample mean is different from the population mean.")
```

2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.

```python
import numpy as np
from scipy import stats

# Simulate two sets of random data (e.g., for a two-sample t-test)
np.random.seed(42) # for reproducibility
data_group1 = np.random.normal(loc=50, scale=10, size=100)
data_group2 = np.random.normal(loc=52, scale=10, size=100)

# Perform an independent t-test (a common hypothesis test)
# H0: The means of the two groups are equal.
# Ha: The means of the two groups are not equal.
t_statistic, p_value = stats.ttest_ind(data_group1, data_group2)

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the P-value
alpha = 0.05
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("There is a statistically significant difference between the means of the two groups.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("There is no statistically significant difference between the means of the two groups.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats.weightstats import ztest

# Parameters
sample_mean = 72
sample_std = 10
sample_size = 50
population_mean = 70
alpha = 0.05 # Significance level

# Simulate sample data
sample_data = np.random.normal(sample_mean, sample_std, sample_size)

# Perform two-tailed Z-test
z_statistic, p_value = ztest(x1=sample_data, value=population_mean)

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Calculate critical Z-values for a two-tailed test
z_critical_lower = norm.ppf(alpha / 2)
z_critical_upper = norm.ppf(1 - alpha / 2)

print(f"Critical Z-values: {z_critical_lower:.4f} and {z_critical_upper:.4f}")

# Plotting the Z-distribution and decision regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution (Z)')
plt.fill_between(x, 0, y, where=(x < z_critical_lower) | (x > z_critical_upper), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_critical_lower, color='red', linestyle='--', label=f'Critical Z = {z_critical_lower:.2f}')
plt.axvline(z_critical_upper, color='red', linestyle='--', label=f'Critical Z = {z_critical_upper:.2f}')
plt.axvline(z_statistic, color='blue', linestyle='-', label=f'Calculated Z = {z_statistic:.2f}')
plt.title('Two-tailed Z-test: Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Interpret the results
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis.")
    print(f"The calculated Z-statistic ({z_statistic:.4f}) falls in the rejection region.")
else:
    print("Conclusion: Fail to reject the null hypothesis.")
    print(f"The calculated Z-statistic ({z_statistic:.4f}) falls outside the rejection region.")
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu_null, mu_alt, sigma, n, alpha):
    """
    Calculates and visualizes Type 1 and Type 2 errors for a one-tailed Z-test.

    Args:
        mu_null (float): Population mean under the null hypothesis.
        mu_alt (float): True population mean under the alternative hypothesis.
        sigma (float): Population standard deviation.
        n (int): Sample size.
        alpha (float): Significance level (Type 1 error probability).
    """
    std_error = sigma / np.sqrt(n)

    # Critical value (for a one-tailed upper test)
    # If H1: mu > mu_null
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu_null + z_critical * std_error

    # Type 1 Error (Alpha): Probability of rejecting H0 when H0 is true
    # This is the area under the H0 distribution to the right of x_critical
    type_1_error = 1 - norm.cdf(z_critical) # By definition, this is alpha

    # Type 2 Error (Beta): Probability of failing to reject H0 when H1 is true
    # This is the area under the H1 distribution to the left of x_critical
    z_beta = (x_critical - mu_alt) / std_error
    type_2_error = norm.cdf(z_beta)

    # Power (1 - Beta): Probability of correctly rejecting H0 when H1 is true
    power = 1 - type_2_error

    print(f"Hypothesized Null Mean (H0): {mu_null}")
    print(f"True Alternative Mean (H1): {mu_alt}")
    print(f"Population Standard Deviation: {sigma}")
    print(f"Sample Size: {n}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Type 1 Error (Alpha): {type_1_error:.4f}")
    print(f"Type 2 Error (Beta): {type_2_error:.4f}")
    print(f"Power (1 - Beta): {power:.4f}")

    # Visualization
    x = np.linspace(min(mu_null, mu_alt) - 3 * std_error, max(mu_null, mu_alt) + 3 * std_error, 500)

    pdf_null = norm.pdf(x, mu_null, std_error)
    pdf_alt = norm.pdf(x, mu_alt, std_error)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf_null, label='Null Hypothesis Distribution (H0)', color='blue')
    plt.plot(x, pdf_alt, label='Alternative Hypothesis Distribution (H1)', color='green')

    # Shade Type 1 Error (alpha)
    x_alpha = np.linspace(x_critical, x.max(), 100)
    plt.fill_between(x_alpha, 0, norm.pdf(x_alpha, mu_null, std_error), color='red', alpha=0.3, label='Type 1 Error (Alpha)')
    plt.axvline(x_critical, color='red', linestyle='--', label=f'Critical Value = {x_critical:.2f}')

    # Shade Type 2 Error (beta)
    x_beta = np.linspace(x.min(), x_critical, 100)
    plt.fill_between(x_beta, 0, norm.pdf(x_beta, mu_alt, std_error), color='orange', alpha=0.3, label='Type 2 Error (Beta)')

    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Usage:
visualize_errors(mu_null=100, mu_alt=105, sigma=15, n=30, alpha=0.05)
```

6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats

# Simulate data for two independent groups
np.random.seed(42) # for reproducibility
group1_data = np.random.normal(loc=60, scale=10, size=50) # e.g., scores from group A
group2_data = np.random.normal(loc=65, scale=10, size=50) # e.g., scores from group B

print(f"Mean of Group 1: {np.mean(group1_data):.2f}")
print(f"Mean of Group 2: {np.mean(group2_data):.2f}")

# Perform independent two-sample t-test
# H0: The means of the two groups are equal (mu1 = mu2)
# Ha: The means of the two groups are not equal (mu1 != mu2)
# `equal_var=True` assumes equal population variances (Student's t-test)
# `equal_var=False` performs Welch's t-test (does not assume equal variances, generally preferred)
t_statistic, p_value = stats.ttest_ind(group1_data, group2_data, equal_var=False)

print(f"\nT-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print(f"\nSince the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference between the means of the two independent groups.")
else:
    print(f"\nSince the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference between the means of the two independent groups.")
```

7. Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulate paired data (e.g., 'before' and 'after' measurements for the same subjects)
np.random.seed(42) # for reproducibility
before_treatment = np.random.normal(loc=70, scale=8, size=30)
after_treatment = before_treatment + np.random.normal(loc=3, scale=2, size=30) # Effect of treatment

print(f"Mean 'Before': {np.mean(before_treatment):.2f}")
print(f"Mean 'After': {np.mean(after_treatment):.2f}")

# Perform paired samples t-test
# H0: The mean difference between 'before' and 'after' is zero (mu_diff = 0)
# Ha: The mean difference is not zero (mu_diff != 0)
t_statistic, p_value = stats.ttest_rel(before_treatment, after_treatment)

print(f"\nT-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print(f"\nSince the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference between the 'before' and 'after' measurements.")
else:
    print(f"\nSince the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference between the 'before' and 'after' measurements.")

# Visualization
plt.figure(figsize=(8, 6))
plt.boxplot([before_treatment, after_treatment], labels=['Before Treatment', 'After Treatment'], patch_artist=True)
plt.title('Comparison of Before and After Treatment Scores (Paired T-test)')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add individual data points (optional, but can be helpful)
for i, data in enumerate([before_treatment, after_treatment]):
    y = data
    x = np.random.normal(i + 1, 0.04, len(y)) # Add jitter for better visibility
    plt.plot(x, y, 'o', color='black', alpha=0.4)

plt.show()

# Another common visualization for paired data: a scatter plot with a line for each pair
plt.figure(figsize=(8, 6))
plt.plot([1] * len(before_treatment), before_treatment, 'o', color='blue', alpha=0.6, label='Before')
plt.plot([2] * len(after_treatment), after_treatment, 'o', color='green', alpha=0.6, label='After')
for i in range(len(before_treatment)):
    plt.plot([1, 2], [before_treatment[i], after_treatment[i]], 'k-', alpha=0.3)
plt.xticks([1, 2], ['Before', 'After'])
plt.title('Paired Samples Comparison')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Args:
        data (array-like): The sample data.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).

    Returns:
        tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1) # Use ddof=1 for sample standard deviation
    n = len(data)

    # Calculate the t-score for the given confidence level and degrees of freedom
    # For a (1-alpha)% confidence interval, we need the t-value such that alpha/2 is in each tail.
    # So, we use (1 - (1 - confidence_level) / 2) for the percentile.
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n - 1)

    # Calculate the margin of error
    margin_of_error = t_critical * (sample_std / np.sqrt(n))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean, margin_of_error

# Example Usage:
sample_data = np.array([85, 90, 78, 92, 88, 76, 89, 81, 95, 83, 87, 80])
confidence_level = 0.95

lower, upper, mean, moe = calculate_confidence_interval(sample_data, confidence_level)

print(f"Sample Mean: {mean:.2f}")
print(f"Confidence Level: {confidence_level * 100}%")
print(f"Margin of Error: {moe:.2f}")
print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")

print("\n--- Significance of Confidence Interval ---")
print("A confidence interval provides a range of values within which the true population mean is likely to lie.")
print(f"For example, a {confidence_level * 100}% confidence interval means that if we were to take many samples and construct a confidence interval for each, approximately {confidence_level * 100}% of those intervals would contain the true population mean.")
print("It quantifies the uncertainty around the sample mean as an estimate of the population mean.")
print("A narrower interval suggests a more precise estimate, while a wider interval suggests more uncertainty.")
print("Confidence intervals are useful for:")
print("- Estimating population parameters (e.g., mean, proportion).")
print("- Comparing groups: If the confidence intervals for two groups do not overlap, it suggests a statistically significant difference between their population means.")
print("- Providing a range of plausible values rather than a single point estimate.")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence_level=0.95):
    """
    Calculates the margin of error for a sample mean.

    Args:
        data (array-like): The sample data.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).

    Returns:
        float: The margin of error.
    """
    sample_std = np.std(data, ddof=1) # Sample standard deviation
    n = len(data)

    # For a given confidence level, we need the t-score (or z-score if n is very large and pop std known)
    # For a (1-alpha)% confidence interval, we need the t-value such that alpha/2 is in each tail.
    # So, we use (1 - (1 - confidence_level) / 2) for the percentile.
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n - 1)

    margin_of_error = t_critical * (sample_std / np.sqrt(n))
    return margin_of_error

# Example Usage:
sample_scores = np.array([75, 80, 85, 70, 90, 65, 95, 78, 82, 88])
confidence_level = 0.99 # e.g., for a 99% confidence interval

moe = calculate_margin_of_error(sample_scores, confidence_level)

print(f"Sample Data: {sample_scores}")
print(f"Sample Mean: {np.mean(sample_scores):.2f}")
print(f"Sample Standard Deviation: {np.std(sample_scores, ddof=1):.2f}")
print(f"Sample Size: {len(sample_scores)}")
print(f"Confidence Level: {confidence_level * 100}%")
print(f"Calculated Margin of Error: {moe:.4f}")

# To illustrate its effect on the confidence interval:
sample_mean = np.mean(sample_scores)
lower_bound = sample_mean - moe
upper_bound = sample_mean + moe
print(f"Corresponding Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

print("\n--- How Margin of Error Affects the Confidence Interval ---")
print("The margin of error (MOE) is the 'plus-minus' value that defines the width of the confidence interval.")
print("Confidence Interval = Sample Mean ± Margin of Error")
print("A larger margin of error leads to a wider confidence interval, indicating more uncertainty in our estimate of the population mean.")
print("A smaller margin of error leads to a narrower confidence interval, indicating a more precise estimate.")
print("\nThe margin of error is influenced by:")
print("- **Confidence Level:** A higher confidence level (e.g., 99% vs. 95%) requires a larger critical value (t-score or z-score), thus increasing the MOE and widening the interval.")
print("- **Sample Standard Deviation:** A larger sample standard deviation (more variability in the data) leads to a larger MOE.")
print("- **Sample Size:** A larger sample size leads to a smaller standard error (sample_std / sqrt(n)), thus decreasing the MOE and narrowing the interval. This is because larger samples provide more information about the population, leading to more precise estimates.")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
import numpy as np

def bayesian_inference_coin_toss(num_flips, num_heads, prior_alpha=1, prior_beta=1):
    """
    Implements Bayesian inference for estimating the probability of heads in a coin toss
    using a Beta-Binomial conjugate prior-posterior relationship.

    Args:
        num_flips (int): Total number of coin flips (observations).
        num_heads (int): Number of heads observed.
        prior_alpha (float): Alpha parameter for the Beta prior distribution.
                              Represents 'pseudo-counts' of heads in prior belief.
        prior_beta (float): Beta parameter for the Beta prior distribution.
                             Represents 'pseudo-counts' of tails in prior belief.

    Returns:
        tuple: (posterior_alpha, posterior_beta) parameters of the Beta posterior distribution.
    """
    # Bayes' Theorem: P(H|E) = P(E|H) * P(H) / P(E)
    # In the context of Beta-Binomial:
    # Prior: P(theta) ~ Beta(prior_alpha, prior_beta)
    # Likelihood: P(data | theta) ~ Binomial(num_flips, theta)
    # Posterior: P(theta | data) ~ Beta(prior_alpha + num_heads, prior_beta + num_tails)

    num_tails = num_flips - num_heads

    # Update prior parameters to get posterior parameters
    posterior_alpha = prior_alpha + num_heads
    posterior_beta = prior_beta + num_tails

    return posterior_alpha, posterior_beta

# --- Explanation of the Bayesian Inference Process ---

print("--- Bayesian Inference Process (Coin Toss Example) ---")

print("\n**1. Define the Prior Distribution (P(H))**")
print("The prior distribution represents our initial belief about the parameter (e.g., probability of heads, 'theta') before observing any data.")
print("For a coin toss, a common choice for the prior is the Beta distribution because it's conjugate to the Binomial likelihood, making calculations tractable.")
print("Beta(alpha, beta) can represent pseudo-counts of heads (alpha-1) and tails (beta-1).")
print("A common 'uninformative' prior is Beta(1,1), which is a uniform distribution, meaning all probabilities are equally likely initially.")

prior_alpha_ex = 1  # Equivalent to assuming 0 heads and 0 tails initially
prior_beta_ex = 1   # (Beta(1,1) is a uniform prior)
print(f"Example Prior: Beta({prior_alpha_ex}, {prior_beta_ex}) (Uniform Prior)")

print("\n**2. Collect Data / Define the Likelihood (P(E|H))**")
print("The likelihood function describes the probability of observing the data given a specific value of the parameter.")
print("For a coin toss, the number of heads in a fixed number of flips follows a Binomial distribution.")
print("Likelihood: P(Number of Heads | theta, Number of Flips)")

num_flips_ex = 10
num_heads_ex = 7
print(f"Example Data: {num_flips_ex} flips, {num_heads_ex} heads")

print("\n**3. Calculate the Posterior Distribution (P(H|E)) using Bayes' Theorem**")
print("Bayes' Theorem updates our prior belief based on the observed data to produce the posterior distribution.")
print("Posterior = (Likelihood * Prior) / Evidence")
print("For Beta-Binomial, the posterior is also a Beta distribution with updated parameters:")
print("new_alpha = prior_alpha + number_of_heads")
print("new_beta = prior_beta + number_of_tails")

posterior_alpha_result, posterior_beta_result = bayesian_inference_coin_toss(
    num_flips=num_flips_ex,
    num_heads=num_heads_ex,
    prior_alpha=prior_alpha_ex,
    prior_beta=prior_beta_ex
)

print(f"Calculated Posterior: Beta({posterior_alpha_result}, {posterior_beta_result})")

# --- Interpretation of the Posterior ---
print("\n**4. Interpret the Posterior Distribution**")
print("The posterior distribution represents our updated belief about the parameter after incorporating the data.")
print("The mean of a Beta(alpha, beta) distribution is alpha / (alpha + beta). This is our point estimate for the probability of heads.")
posterior_mean = posterior_alpha_result / (posterior_alpha_result + posterior_beta_result)
print(f"Posterior Mean (Estimated Probability of Heads): {posterior_mean:.4f}")

print("\nWe can also calculate a credible interval from the posterior, which is the Bayesian equivalent of a confidence interval.")
# For example, a 95% credible interval
credible_interval_lower, credible_interval_upper = stats.beta.interval(
    0.95, posterior_alpha_result, posterior_beta_result
)
print(f"95% Credible Interval for Probability of Heads: ({credible_interval_lower:.4f}, {credible_interval_upper:.4f})")

print("\n--- Significance of Bayesian Inference ---")
print("- Allows incorporating prior knowledge into the analysis.")
print("- Provides a full probability distribution for the parameter, not just a point estimate.")
print("- Naturally handles small sample sizes and updates beliefs iteratively as more data becomes available.")
print("- Credible intervals are more intuitive than frequentist confidence intervals: 'There is a 95% probability that the true parameter lies within this interval.'")
print("- It's particularly useful in situations where prior information is strong or when frequentist methods are difficult to apply.")

# Example with a different prior (e.g., biased coin expected)
print("\n--- Example with a more 'informative' prior (expecting a slightly biased coin) ---")
# Prior belief that coin is slightly biased towards heads, e.g., Beta(5, 2)
# Mean of this prior is 5/(5+2) = 0.71
prior_alpha_biased = 5
prior_beta_biased = 2
print(f"Prior: Beta({prior_alpha_biased}, {prior_beta_biased}) (Mean: {prior_alpha_biased/(prior_alpha_biased + prior_beta_biased):.2f})")

posterior_alpha_biased, posterior_beta_biased = bayesian_inference_coin_toss(
    num_flips=num_flips_ex,
    num_heads=num_heads_ex,
    prior_alpha=prior_alpha_biased,
    prior_beta=prior_beta_biased
)
print(f"Posterior: Beta({posterior_alpha_biased}, {posterior_beta_biased})")
print(f"Posterior Mean (Estimated Probability of Heads): {posterior_alpha_biased / (posterior_alpha_biased + posterior_beta_biased):.4f}")
print("Notice how the prior influences the posterior, especially with limited data.")
```

12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
from scipy.stats import chi2_contingency
import pandas as pd

# Create a contingency table (observed frequencies)
# Example: Relationship between Gender and Preferred Sport
# Rows: Gender (Male, Female)
# Columns: Preferred Sport (Football, Basketball, Tennis)

# Data example:
#               Football  Basketball  Tennis
# Male                30          20      10
# Female              15          25      30
data = {
    'Football': [30, 15],
    'Basketball': [20, 25],
    'Tennis': [10, 30]
}
contingency_table = pd.DataFrame(data, index=['Male', 'Female'])

print("Observed Contingency Table:")
print(contingency_table)

# Perform the Chi-square test for independence
# H0: The two categorical variables are independent.
# Ha: The two categorical variables are dependent (related).
chi2, p_value, degrees_of_freedom, expected_frequencies = chi2_contingency(contingency_table)

print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print("\nExpected Frequencies (under the assumption of independence):")
print(pd.DataFrame(expected_frequencies, index=['Male', 'Female'], columns=['Football', 'Basketball', 'Tennis']))

# Interpret the results
alpha = 0.05
print("\n--- Interpretation ---")
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant association (dependence) between Gender and Preferred Sport.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant association (independence) between Gender and Preferred Sport.")

print("\nExplanation:")
print("The Chi-square test for independence determines if there's a relationship between two categorical variables.")
print("It compares the observed frequencies in each cell of the contingency table to the frequencies that would be expected if the variables were truly independent.")
print("A large Chi-square statistic and a small P-value indicate that the observed frequencies are significantly different from the expected frequencies, suggesting dependence.")
```

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed_data):
    """
    Calculates the expected frequencies for a Chi-square test of independence
    based on the observed contingency table.

    Args:
        observed_data (numpy.ndarray or pandas.DataFrame): The observed
                                                           contingency table.

    Returns:
        numpy.ndarray: The expected frequencies table.
    """
    observed_array = np.array(observed_data)
    row_sums = observed_array.sum(axis=1, keepdims=True)
    col_sums = observed_array.sum(axis=0, keepdims=True)
    total_sum = observed_array.sum()

    expected_frequencies = (row_sums * col_sums) / total_sum
    return expected_frequencies

# Example Observed Data (as a pandas DataFrame for clarity)
# This is the same example from question 12
data = {
    'Football': [30, 15],
    'Basketball': [20, 25],
    'Tennis': [10, 30]
}
observed_contingency_table = pd.DataFrame(data, index=['Male', 'Female'])

print("Observed Contingency Table:")
print(observed_contingency_table)

# Calculate expected frequencies
expected_freq_array = calculate_expected_frequencies(observed_contingency_table)

# Convert to DataFrame for better display with original labels
expected_contingency_table = pd.DataFrame(expected_freq_array,
                                          index=observed_contingency_table.index,
                                          columns=observed_contingency_table.columns)

print("\nCalculated Expected Frequencies Table:")
print(expected_contingency_table.round(2)) # Round for cleaner display

# Verification (using scipy's chi2_contingency to compare)
from scipy.stats import chi2_contingency
_chi2, _p, _dof, expected_from_scipy = chi2_contingency(observed_contingency_table)

print("\nExpected Frequencies from scipy.stats.chi2_contingency (for comparison):")
print(pd.DataFrame(expected_from_scipy,
                   index=observed_contingency_table.index,
                   columns=observed_contingency_table.columns).round(2))

print("\n--- Explanation of Expected Frequencies ---")
print("Expected frequencies represent the cell counts that would be anticipated if the two categorical variables were perfectly independent.")
print("The formula for an expected frequency in a cell is: (Row Total * Column Total) / Grand Total.")
print("These expected frequencies are crucial for the Chi-square test, as the test statistic is calculated by comparing the observed frequencies to these expected frequencies. A large deviation between observed and expected suggests dependence.")
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
from scipy.stats import chisquare

# Example: M&M's colors distribution
# According to a hypothetical claim, M&M's Plain candies have the following color distribution:
# Blue: 24%, Orange: 20%, Green: 16%, Yellow: 14%, Red: 13%, Brown: 13%

# Observed data from a bag of 100 M&M's
observed_counts = np.array([20, 22, 18, 10, 15, 15]) # Blue, Orange, Green, Yellow, Red, Brown
total_observed = np.sum(observed_counts)

# Expected proportions
expected_proportions = np.array([0.24, 0.20, 0.16, 0.14, 0.13, 0.13])

# Calculate expected counts based on total observed
expected_counts = expected_proportions * total_observed

print(f"Observed Counts: {observed_counts}")
print(f"Expected Proportions: {expected_proportions}")
print(f"Calculated Expected Counts (based on {total_observed} total observed): {expected_counts.round(2)}")

# Perform the Chi-square Goodness-of-Fit test
# H0: The observed data fits the expected distribution.
# Ha: The observed data does not fit the expected distribution.
chi2_statistic, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

print(f"\nChi-square Statistic (Goodness-of-Fit): {chi2_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
print("\n--- Interpretation ---")
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: The observed M&M's color distribution does not significantly fit the expected distribution.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: The observed M&M's color distribution significantly fits the expected distribution (or there's no strong evidence to suggest it doesn't).")

print("\nExplanation:")
print("The Chi-square goodness-of-fit test is used to determine if an observed frequency distribution differs significantly from an expected frequency distribution.")
print("It's commonly used to test if a sample comes from a specific population distribution (e.g., uniform, normal, or a specified categorical distribution).")
print("The test calculates a statistic that measures the discrepancy between observed and expected counts. A larger discrepancy leads to a larger chi-square value and a smaller P-value, suggesting a poor fit.")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

def f_test_variances(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two independent samples.

    H0: The variances of the two populations are equal (sigma1^2 = sigma2^2)
    Ha: The variances of the two populations are not equal (sigma1^2 != sigma2^2)

    Args:
        sample1 (array-like): Data from the first sample.
        sample2 (array-like): Data from the second sample.
        alpha (float): Significance level.

    Returns:
        tuple: (f_statistic, p_value, interpretation)
    """
    n1 = len(sample1)
    n2 = len(sample2)

    # Calculate sample variances (unbiased estimate with ddof=1)
    s1_squared = np.var(sample1, ddof=1)
    s2_squared = np.var(sample2, ddof=1)

    print(f"Sample 1 Size: {n1}, Sample 1 Variance: {s1_squared:.4f}")
    print(f"Sample 2 Size: {n2}, Sample 2 Variance: {s2_squared:.4f}")

    # Ensure F-statistic is always >= 1 for easier lookup in F-table or p-value calculation
    # F = (Larger Variance) / (Smaller Variance)
    if s1_squared >= s2_squared:
        f_statistic = s1_squared / s2_squared
        dfn = n1 - 1 # Degrees of freedom numerator
        dfd = n2 - 1 # Degrees of freedom denominator
    else:
        f_statistic = s2_squared / s1_squared
        dfn = n2 - 1
        dfd = n1 - 1

    # Calculate the p-value for a two-tailed test
    # Since we made F >= 1, we consider the right tail and then multiply by 2 for two-tailed
    p_value = 2 * (1 - f.cdf(f_statistic, dfn, dfd))

    print(f"\nCalculated F-statistic: {f_statistic:.4f}")
    print(f"Degrees of Freedom (numerator, denominator): ({dfn}, {dfd})")
    print(f"P-value: {p_value:.4f}")

    # Interpret the results
    if p_value < alpha:
        interpretation = f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.\nConclusion: There is a statistically significant difference between the variances of the two populations."
    else:
        interpretation = f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.\nConclusion: There is no statistically significant difference between the variances of the two populations."

    return f_statistic, p_value, interpretation

# Example Usage:
np.random.seed(42)
sample_A = np.random.normal(loc=10, scale=3, size=25) # Variance = 3^2 = 9
sample_B = np.random.normal(loc=12, scale=5, size=30) # Variance = 5^2 = 25

f_stat, p_val, interp = f_test_variances(sample_A, sample_B, alpha=0.05)
print(f"\n{interp}")

print("\n--- Another Example (variances are likely similar) ---")
sample_C = np.random.normal(loc=50, scale=7, size=40)
sample_D = np.random.normal(loc=55, scale=7.5, size=35)

f_stat_2, p_val_2, interp_2 = f_test_variances(sample_C, sample_D, alpha=0.05)
print(f"\n{interp_2}")

print("\n--- Significance of F-test for Variances ---")
print("The F-test for variances is used to determine if two population variances are equal.")
print("This test is often a prerequisite for other statistical tests, such as the independent samples t-test.")
print("If the F-test indicates unequal variances, Welch's t-test (which doesn't assume equal variances) should be used instead of Student's t-test.")
print("It's sensitive to departures from normality, especially with small sample sizes.")
```

17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

```python
import numpy as np
from scipy.stats import f_oneway
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data for three different groups
np.random.seed(42) # for reproducibility

# Group 1: e.g., scores from Teaching Method A
group1 = np.random.normal(loc=75, scale=10, size=30)
# Group 2: e.g., scores from Teaching Method B
group2 = np.random.normal(loc=80, scale=10, size=30)
# Group 3: e.g., scores from Teaching Method C
group3 = np.random.normal(loc=72, scale=10, size=30)

print(f"Mean Group 1: {np.mean(group1):.2f}")
print(f"Mean Group 2: {np.mean(group2):.2f}")
print(f"Mean Group 3: {np.mean(group3):.2f}")

# Perform One-Way ANOVA test
# H0: The means of all groups are equal (mu1 = mu2 = mu3)
# Ha: At least one group mean is different from the others
f_statistic, p_value = f_oneway(group1, group2, group3)

print(f"\nF-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
print("\n--- Interpretation ---")
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference between the means of at least two of the groups.")
    print("Note: ANOVA tells us *if* there's a difference, but not *which* specific groups differ. Post-hoc tests (e.g., Tukey HSD) would be needed for that.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference between the means of the groups.")

# Visualization (using boxplots)
data_for_plot = pd.DataFrame({
    'Score': np.concatenate([group1, group2, group3]),
    'Group': ['Group 1']*len(group1) + ['Group 2']*len(group2) + ['Group 3']*len(group3)
})

plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Score', data=data_for_plot)
plt.title('Comparison of Group Means (ANOVA)')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("\n--- Assumptions of ANOVA ---")
print("For ANOVA to be valid, several assumptions should ideally be met:")
print("1.  **Independence of Observations:** The observations within and between groups must be independent.")
print("2.  **Normality:** The data in each group should be approximately normally distributed.")
print("    - ANOVA is relatively robust to minor departures from normality, especially with larger sample sizes.")
print("3.  **Homogeneity of Variances (Homoscedasticity):** The variances of the populations from which the samples are drawn should be approximately equal.")
print("    - This can be checked using tests like Levene's test or Bartlett's test.")
print("    - If this assumption is violated, Welch's ANOVA can be used (which doesn't assume equal variances).")
print("4.  **Interval/Ratio Data:** The dependent variable (scores in this case) should be measured on an interval or ratio scale.")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

```python
import numpy as np
from scipy import stats
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import lilliefors, het_breuschpagan, het_goldfeldquandt

def check_anova_assumptions(data_frame, dependent_var, group_var):
    """
    Checks the assumptions for a one-way ANOVA:
    1. Normality (Shapiro-Wilk test) for each group.
    2. Independence (assumed by proper experimental design, not statistically tested here).
    3. Homogeneity of Variances (Levene's test).

    Args:
        data_frame (pd.DataFrame): DataFrame containing the data.
        dependent_var (str): Name of the dependent (continuous) variable column.
        group_var (str): Name of the grouping (categorical) variable column.
    """
    print("--- Checking ANOVA Assumptions ---")

    groups = data_frame[group_var].unique()

    # 1. Normality Assumption (Shapiro-Wilk Test for each group)
    print("\n1. Normality Check (Shapiro-Wilk Test):")
    print("   H0: The data is drawn from a normal distribution.")
    print("   Ha: The data is not drawn from a normal distribution.")
    normality_ok = True
    for group_name in groups:
        group_data = data_frame[data_frame[group_var] == group_name][dependent_var]
        if len(group_data) > 3: # Shapiro-Wilk requires at least 4 data points
            stat, p = stats.shapiro(group_data)
            print(f"   - Group '{group_name}': Statistic={stat:.4f}, P-value={p:.4f}")
            if p < 0.05: # Using alpha = 0.05 for normality
                print(f"     -> P-value < 0.05, suggesting non-normality for group '{group_name}'.")
                normality_ok = False
        else:
            print(f"   - Group '{group_name}': Not enough data points ({len(group_data)}) for Shapiro-Wilk test. Consider visual checks (histograms/Q-Q plots).")

    if normality_ok:
        print("   Conclusion: Normality assumption appears to be reasonably met for all groups (or no strong evidence against it).")
    else:
        print("   Conclusion: Normality assumption may be violated for one or more groups. Consider transformations or non-parametric tests (e.g., Kruskal-Wallis).")

    # 2. Independence of Observations (Conceptual check, not statistical test)
    print("\n2. Independence of Observations:")
    print("   This assumption requires that observations within each group and between groups are independent.")
    print("   This is usually ensured by proper experimental design (e.g., random sampling, no repeated measures on the same subjects unless using repeated measures ANOVA).")
    print("   This assumption cannot be statistically tested from the data itself; it relies on how the data was collected.")

    # 3. Homogeneity of Variances (Levene's Test)
    print("\n3. Homogeneity of Variances Check (Levene's Test):")
    print("   H0: The variances are equal across all groups (homoscedasticity).")
    print("   Ha: At least one group's variance is different (heteroscedasticity).")
    # Levene's test takes positional arguments for each group's data
    group_data_list = [data_frame[data_frame[group_var] == g][dependent_var] for g in groups]
    levene_stat, levene_p = stats.levene(*group_data_list)
    print(f"   - Levene's Test: Statistic={levene_stat:.4f}, P-value={levene_p:.4f}")

    if levene_p < 0.05: # Using alpha = 0.05 for equal variance
        print("     -> P-value < 0.05, suggesting violation of homogeneity of variances.")
        print("     Consider using Welch's ANOVA or data transformation if this assumption is violated.")
    else:
        print("     -> P-value >= 0.05, suggesting homogeneity of variances assumption is met.")

# Example Usage:
# Simulate some data
np.random.seed(42)
data_group1 = np.random.normal(loc=75, scale=10, size=30)
data_group2 = np.random.normal(loc=80, scale=12, size=30) # Slightly different std
data_group3 = np.random.normal(loc=72, scale=9, size=30)

# Create a DataFrame
data_anova = pd.DataFrame({
    'Score': np.concatenate([data_group1, data_group2, data_group3]),
    'Group': ['A']*len(data_group1) + ['B']*len(data_group2) + ['C']*len(data_group3)
})

# Check the assumptions
check_anova_assumptions(data_anova, 'Score', 'Group')

# Example with a group clearly not normal (e.g., from an exponential distribution)
print("\n\n--- Example with a non-normal group and unequal variances ---")
data_group_non_normal = np.random.exponential(scale=10, size=30)
data_group_high_variance = np.random.normal(loc=60, scale=20, size=30) # High variance

data_anova_problematic = pd.DataFrame({
    'Score': np.concatenate([data_group1, data_group_non_normal, data_group_high_variance]),
    'Group': ['A']*len(data_group1) + ['D']*len(data_group_non_normal) + ['E']*len(data_group_high_variance)
})
check_anova_assumptions(data_anova_problematic, 'Score', 'Group')
```

20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data for a two-way ANOVA
np.random.seed(42)

# Factors:
# Factor A: Drug Type (Drug_X, Drug_Y)
# Factor B: Dosage (Low, High)
# Dependent Variable: Recovery Time

# Number of subjects per condition
n_per_group = 20

# Create data for each combination of factors
# Drug_X, Low Dosage
drug_x_low = np.random.normal(loc=10, scale=2, size=n_per_group)
# Drug_X, High Dosage
drug_x_high = np.random.normal(loc=8, scale=2, size=n_per_group) # Drug_X might be more effective at high dosage

# Drug_Y, Low Dosage
drug_y_low = np.random.normal(loc=12, scale=2, size=n_per_group)
# Drug_Y, High Dosage
drug_y_high = np.random.normal(loc=11, scale=2, size=n_per_group) # Drug_Y less sensitive to dosage, or slightly worse at high

# Combine all data into a pandas DataFrame
data = pd.DataFrame({
    'Recovery_Time': np.concatenate([drug_x_low, drug_x_high, drug_y_low, drug_y_high]),
    'Drug_Type': ['Drug_X'] * (2 * n_per_group) + ['Drug_Y'] * (2 * n_per_group),
    'Dosage': ['Low'] * n_per_group + ['High'] * n_per_group + ['Low'] * n_per_group + ['High'] * n_per_group
})

print("Data Head:\n", data.head())
print("\nMean Recovery Time by Drug Type and Dosage:")
print(data.groupby(['Drug_Type', 'Dosage'])['Recovery_Time'].mean().unstack())

# Perform Two-Way ANOVA
# Formula: dependent_var ~ C(factor1) + C(factor2) + C(factor1):C(factor2)
# C() indicates categorical variables
# : indicates interaction term
model = ols('Recovery_Time ~ C(Drug_Type) + C(Dosage) + C(Drug_Type):C(Dosage)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2) # Type 2 sums of squares recommended for unbalanced designs, but often fine for balanced too

print("\n--- Two-Way ANOVA Results ---")
print(anova_table)

# Interpret the results
alpha = 0.05
print("\n--- Interpretation of Two-Way ANOVA ---")
print("1.  **Interaction Effect (C(Drug_Type):C(Dosage))**: This is often the most important part.")
if anova_table['PR(>F)'][2] < alpha: # Index 2 is for the interaction term
    print(f"    - P-value for interaction ({anova_table['PR(>F)'][2]:.4f}) is less than {alpha}.")
    print("    - Conclusion: There is a statistically significant interaction effect between Drug Type and Dosage.")
    print("      This means the effect of Dosage on Recovery Time depends on the type of Drug, AND/OR the effect of Drug Type depends on the Dosage.")
    print("      When interaction is significant, main effects should be interpreted with caution, as the overall effect of one factor might be misleading without considering the other.")
else:
    print(f"    - P-value for interaction ({anova_table['PR(>F)'][2]:.4f}) is greater than {alpha}.")
    print("    - Conclusion: There is no statistically significant interaction effect between Drug Type and Dosage.")
    print("      This suggests the effect of one factor is consistent across the levels of the other factor. We can then proceed to interpret the main effects.")

print("\n2.  **Main Effect of Drug Type (C(Drug_Type))**:")
if anova_table['PR(>F)'][0] < alpha: # Index 0 for Drug_Type
    print(f"    - P-value for Drug Type ({anova_table['PR(>F)'][0]:.4f}) is less than {alpha}. Significant difference between drug types.")
else:
    print(f"    - P-value for Drug Type ({anova_table['PR(>F)'][0]:.4f}) is greater than {alpha}. No significant difference between drug types.")

print("\n3.  **Main Effect of Dosage (C(Dosage))**:")
if anova_table['PR(>F)'][1] < alpha: # Index 1 for Dosage
    print(f"    - P-value for Dosage ({anova_table['PR(>F)'][1]:.4f}) is less than {alpha}. Significant difference between dosages.")
else:
    print(f"    - P-value for Dosage ({anova_table['PR(>F)'][1]:.4f}) is greater than {alpha}. No significant difference between dosages.")

# Visualization of Interaction Effect
plt.figure(figsize=(10, 7))
sns.pointplot(x='Dosage', y='Recovery_Time', hue='Drug_Type', data=data,
              errorbar='se', capsize=0.1, palette='viridis') # 'se' for standard error
plt.title('Interaction Plot: Recovery Time by Drug Type and Dosage')
plt.xlabel('Dosage')
plt.ylabel('Mean Recovery Time')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Drug Type')
plt.show()

print("\n--- Explanation of the Interaction Plot ---")
print("The interaction plot shows the mean of the dependent variable (Recovery Time) for each combination of the two factors.")
print("If the lines on the plot are not parallel (they cross or diverge/converge significantly), it visually suggests an interaction effect.")
print("Parallel lines would suggest no interaction, meaning the effect of one factor is consistent across levels of the other.")
print("In this plot:")
print("- If the lines for Drug_X and Drug_Y have different slopes across 'Low' and 'High' dosages, it indicates an interaction.")
print("- If the lines are roughly parallel, it suggests no significant interaction.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Define degrees of freedom for numerator (dfn) and denominator (dfd)
dfs_combinations = [(1, 10), (5, 10), (10, 10), (10, 50)] # (dfn, dfd)

x = np.linspace(0, 5, 500) # Range for F-statistic values

plt.figure(figsize=(10, 6))

for dfn, dfd in dfs_combinations:
    pdf = f.pdf(x, dfn, dfd)
    plt.plot(x, pdf, label=f'dfn={dfn}, dfd={dfd}')

plt.title('F-Distribution for Various Degrees of Freedom')
plt.xlabel('F-statistic Value')
plt.ylabel('Probability Density Function')
plt.legend()
plt.grid(True)
plt.xlim(0, 5) # Adjust x-axis limit for better visualization
plt.ylim(0, 1.0) # Adjust y-axis limit
plt.show()

print("--- Characteristics and Use of the F-Distribution in Hypothesis Testing ---")
print("1.  **Shape:** The F-distribution is non-negative and typically positively skewed (skewed to the right).")
print("    - Its shape is determined by two parameters: degrees of freedom for the numerator (df1 or dfn) and degrees of freedom for the denominator (df2 or dfd).")
print("    - As the degrees of freedom increase, the distribution becomes more symmetrical and bell-shaped, approaching a normal distribution (though it remains positive).")
print("    - It starts at 0 and extends towards positive infinity.")
print("2.  **Parameters (Degrees of Freedom):**")
print("    - **dfn (numerator degrees of freedom):** Related to the number of groups or variables being compared in the numerator of the F-statistic.")
print("    - **dfd (denominator degrees of freedom):** Related to the total number of observations minus the number of groups, representing the variability within groups or error.")
print("3.  **Use in Hypothesis Testing:**")
print("    - The F-distribution is fundamental for tests that compare variances or compare means across multiple groups.")
print("    - **ANOVA (Analysis of Variance):** This is the most common application. ANOVA uses the F-statistic to test the null hypothesis that the means of two or more groups are equal.")
print("        - F-statistic in ANOVA = (Variance Between Groups) / (Variance Within Groups)")
print("        - A large F-statistic suggests that the variance between groups is much larger than the variance within groups, leading to a rejection of the null hypothesis (means are different).")
print("    - **F-test for Equality of Variances:** Used to test if the variances of two populations are equal (as demonstrated in question 16).")
print("    - **Regression Analysis:** Used to test the overall significance of a regression model (whether the model explains a significant portion of the variance in the dependent variable).")
print("4.  **One-tailed Test:** F-tests are almost always one-tailed (right-tailed) because we are usually interested in whether the ratio of variances is significantly *greater* than 1 (i.e., whether the numerator variance is significantly larger). A very small F-statistic (close to zero) simply means the numerator variance is smaller than the denominator, which doesn't lead to rejection of the usual null hypothesis.")
print("5.  **Relationship to Chi-square and T-distribution:**")
print("    - If $X_1 \sim \chi^2(d_1)$ and $X_2 \sim \chi^2(d_2)$ are independent chi-squared random variables, then the ratio $(X_1/d_1) / (X_2/d_2)$ follows an F-distribution with degrees of freedom $(d_1, d_2)$.")
print("    - The square of a t-distributed random variable with 'd' degrees of freedom, $t^2(d)$, follows an F-distribution with degrees of freedom $(1, d)$.")
```

22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

```python
import numpy as np
import pandas as pd
from scipy.stats import f_oneway
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate data for different exercise programs and their effect on weight loss
np.random.seed(42)

# Group A: Standard Exercise Program
group_A_weight_loss = np.random.normal(loc=3.0, scale=1.5, size=50) # Mean 3.0 kg loss
# Group B: High-Intensity Interval Training (HIIT)
group_B_weight_loss = np.random.normal(loc=4.5, scale=1.5, size=50) # Mean 4.5 kg loss
# Group C: Strength Training Only
group_C_weight_loss = np.random.normal(loc=2.5, scale=1.5, size=50) # Mean 2.5 kg loss

# Create a DataFrame for easier plotting
data = {
    'Weight_Loss_kg': np.concatenate([group_A_weight_loss, group_B_weight_loss, group_C_weight_loss]),
    'Program': ['Standard'] * len(group_A_weight_loss) + ['HIIT'] * len(group_B_weight_loss) + ['Strength'] * len(group_C_weight_loss)
}
df = pd.DataFrame(data)

print("Mean Weight Loss for each program:")
print(df.groupby('Program')['Weight_Loss_kg'].mean().round(2))

# Perform One-Way ANOVA
# H0: The mean weight loss is the same across all exercise programs.
# Ha: At least one exercise program leads to different mean weight loss.
f_statistic, p_value = f_oneway(group_A_weight_loss, group_B_weight_loss, group_C_weight_loss)

print(f"\nOne-Way ANOVA F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
print("\n--- Interpretation ---")
if p_value < alpha:
    print(f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a statistically significant difference in mean weight loss among the different exercise programs.")
    print("A post-hoc test (e.g., Tukey HSD) would be necessary to identify which specific pairs of groups differ.")
else:
    print(f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: There is no statistically significant difference in mean weight loss among the different exercise programs.")

# Visualize the results with boxplots
plt.figure(figsize=(9, 6))
sns.boxplot(x='Program', y='Weight_Loss_kg', data=df, palette='pastel')
sns.swarmplot(x='Program', y='Weight_Loss_kg', data=df, color='black', alpha=0.5, size=3) # Add individual data points
plt.title('Mean Weight Loss by Exercise Program')
plt.xlabel('Exercise Program')
plt.ylabel('Weight Loss (kg)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print("\n--- Explanation of Boxplots for Comparing Group Means ---")
print("Boxplots are excellent for visualizing the distribution of a continuous variable across different categorical groups.")
print("Each box represents a group, showing:")
print("- The central line: The median of the data for that group.")
print("- The box: The interquartile range (IQR), representing the middle 50% of the data (from 25th to 75th percentile).")
print("- The 'whiskers': Extend to the minimum and maximum values within 1.5 times the IQR from the box (excluding outliers).")
print("- Dots outside whiskers: Individual outliers.")
print("By visually comparing the boxes, especially their medians and the overlap of their IQRs, we can get an intuitive sense of whether the group means (or medians) are likely to be different. If the boxes (or their central lines) are substantially non-overlapping, it often aligns with a significant ANOVA result.")
```

23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate random data from a normal distribution
np.random.seed(42) # for reproducibility

# Scenario 1: One Sample Z-test (comparing sample mean to a known population mean)
# Assume population mean is known, and population std is known or sample is large
population_mean_h0 = 100
population_std = 15
sample_size_z = 50
sample_data_z = np.random.normal(loc=103, scale=population_std, size=sample_size_z) # True mean is 103

print("--- Scenario 1: One-Sample Z-test ---")
print(f"Sample Mean: {np.mean(sample_data_z):.2f}")
print(f"Hypothesized Population Mean (H0): {population_mean_h0}")

# Perform Z-test
# H0: Sample mean is equal to population_mean_h0
# Ha: Sample mean is not equal to population_mean_h0
z_statistic, p_value_z = stats.ztest(sample_data_z, value=population_mean_h0, std=population_std)

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value_z:.4f}")

alpha = 0.05
if p_value_z < alpha:
    print(f"Conclusion: Reject H0. (P={p_value_z:.4f} < {alpha}) - Sample mean is significantly different from {population_mean_h0}.")
else:
    print(f"Conclusion: Fail to reject H0. (P={p_value_z:.4f} >= {alpha}) - Sample mean is not significantly different from {population_mean_h0}.")

# Visualize the distribution
plt.figure(figsize=(8, 5))
sns.histplot(sample_data_z, kde=True, bins=10)
plt.axvline(np.mean(sample_data_z), color='red', linestyle='--', label=f'Sample Mean: {np.mean(sample_data_z):.2f}')
plt.axvline(population_mean_h0, color='blue', linestyle=':', label=f'H0 Mean: {population_mean_h0}')
plt.title('Distribution of Sample Data (Z-test)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()

# Scenario 2: Independent Two-Sample T-test (comparing two sample means)
# Assume population standard deviations are unknown and possibly unequal (Welch's t-test)
sample_size_t1 = 30
sample_size_t2 = 35
sample_data_t1 = np.random.normal(loc=50, scale=5, size=sample_size_t1) # True mean 50
sample_data_t2 = np.random.normal(loc=53, scale=6, size=sample_size_t2) # True mean 53

print("\n--- Scenario 2: Independent Two-Sample T-test ---")
print(f"Sample 1 Mean: {np.mean(sample_data_t1):.2f}")
print(f"Sample 2 Mean: {np.mean(sample_data_t2):.2f}")

# Perform Independent T-test (Welch's t-test for unequal variances)
# H0: Means are equal (mu1 = mu2)
# Ha: Means are not equal (mu1 != mu2)
t_statistic, p_value_t = stats.ttest_ind(sample_data_t1, sample_data_t2, equal_var=False)

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value_t:.4f}")

if p_value_t < alpha:
    print(f"Conclusion: Reject H0. (P={p_value_t:.4f} < {alpha}) - There is a significant difference between the two sample means.")
else:
    print(f"Conclusion: Fail to reject H0. (P={p_value_t:.4f} >= {alpha}) - There is no significant difference between the two sample means.")

# Visualize the distributions
plt.figure(figsize=(10, 6))
sns.histplot(sample_data_t1, kde=True, color='purple', label='Sample 1', alpha=0.6, bins=10)
sns.histplot(sample_data_t2, kde=True, color='orange', label='Sample 2', alpha=0.6, bins=10)
plt.axvline(np.mean(sample_data_t1), color='purple', linestyle='--', label=f'Mean 1: {np.mean(sample_data_t1):.2f}')
plt.axvline(np.mean(sample_data_t2), color='orange', linestyle='--', label=f'Mean 2: {np.mean(sample_data_t2):.2f}')
plt.title('Distributions of Two Sample Means (T-test)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
from scipy.stats import chi2

def chi_square_test_variance(sample_data, hypothesized_population_variance, alpha=0.05, two_tailed=True):
    """
    Performs a Chi-square test for a single population variance.

    H0: Population variance = hypothesized_population_variance (sigma^2 = sigma0^2)
    Ha (two-tailed): Population variance != hypothesized_population_variance
    Ha (one-tailed, upper): Population variance > hypothesized_population_variance
    Ha (one-tailed, lower): Population variance < hypothesized_population_variance

    Args:
        sample_data (array-like): The sample data.
        hypothesized_population_variance (float): The hypothesized value for the population variance.
        alpha (float): Significance level.
        two_tailed (bool): If True, performs a two-tailed test. Otherwise, assumes an upper-tailed test.
                           (For simplicity, this function implements two-tailed and upper-tailed tests.
                           A lower-tailed test would have a different critical region/p-value calculation.)

    Returns:
        tuple: (chi2_statistic, p_value, interpretation)
    """
    n = len(sample_data)
    sample_variance = np.var(sample_data, ddof=1) # Sample variance (unbiased estimate)

    print(f"Sample Size (n): {n}")
    print(f"Sample Variance (s^2): {sample_variance:.4f}")
    print(f"Hypothesized Population Variance (sigma0^2): {hypothesized_population_variance}")

    # Calculate the Chi-square test statistic
    # chi2_statistic = ((n - 1) * s^2) / sigma0^2
    chi2_statistic = ((n - 1) * sample_variance) / hypothesized_population_variance
    degrees_of_freedom = n - 1

    print(f"\nCalculated Chi-square Statistic: {chi2_statistic:.4f}")
    print(f"Degrees of Freedom: {degrees_of_freedom}")

    # Calculate P-value
    if two_tailed:
        # For a two-tailed test, find the probability in both tails
        p_value_right_tail = 1 - chi2.cdf(chi2_statistic, degrees_of_freedom)
        p_value_left_tail = chi2.cdf(chi2_statistic, degrees_of_freedom)
        p_value = 2 * min(p_value_right_tail, p_value_left_tail) # Take the smaller tail and multiply by 2
        test_type = "Two-tailed"
    else: # One-tailed (upper tail implemented here)
        p_value = 1 - chi2.cdf(chi2_statistic, degrees_of_freedom)
        test_type = "One-tailed (Upper)"

    print(f"P-value ({test_type}): {p_value:.4f}")

    # Interpret the results
    interpretation = ""
    if p_value < alpha:
        interpretation = f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis."
        if two_tailed:
            interpretation += f"\nConclusion: There is significant evidence that the population variance is different from {hypothesized_population_variance}."
        else:
            interpretation += f"\nConclusion: There is significant evidence that the population variance is greater than {hypothesized_population_variance}."
    else:
        interpretation = f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis."
        if two_tailed:
            interpretation += f"\nConclusion: There is no significant evidence that the population variance is different from {hypothesized_population_variance}."
        else:
            interpretation += f"\nConclusion: There is no significant evidence that the population variance is greater than {hypothesized_population_variance}."
    return chi2_statistic, p_value, interpretation

# Example Usage:
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=12, size=20) # True population std = 12, variance = 144
hypothesized_variance = 100 # Test if variance is 100

print("--- Testing if Population Variance = 100 (Two-tailed) ---")
chi2_stat, p_val, interp = chi_square_test_variance(sample_data, hypothesized_variance, alpha=0.05, two_tailed=True)
print(f"\n{interp}")

print("\n\n--- Testing if Population Variance > 100 (One-tailed Upper) ---")
hypothesized_variance_one_tail = 100
chi2_stat_one, p_val_one, interp_one = chi_square_test_variance(sample_data, hypothesized_variance_one_tail, alpha=0.05, two_tailed=False)
print(f"\n{interp_one}")

print("\n--- Significance of Chi-square Test for Population Variance ---")
print("This test is used to assess if a population's variance differs significantly from a hypothesized value.")
print("It assumes that the population from which the sample is drawn is normally distributed.")
print("This test is highly sensitive to departures from normality.")
print("It's less commonly used than tests for means, but is important in quality control, manufacturing, or whenever consistency/variability is a primary concern.")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def z_test_proportions(count1, nobs1, count2, nobs2, alpha=0.05, alternative='two-sided'):
    """
    Performs a Z-test to compare two population proportions.

    H0: p1 = p2 (Proportions are equal)
    Ha:
        'two-sided': p1 != p2
        'smaller': p1 < p2
        'larger': p1 > p2

    Args:
        count1 (int): Number of successes in the first sample.
        nobs1 (int): Total number of observations in the first sample.
        count2 (int): Number of successes in the second sample.
        nobs2 (int): Total number of observations in the second sample.
        alpha (float): Significance level.
        alternative (str): Defines the alternative hypothesis ('two-sided', 'smaller', 'larger').

    Returns:
        tuple: (z_statistic, p_value, interpretation)
    """
    prop1 = count1 / nobs1
    prop2 = count2 / nobs2

    print(f"Sample 1: Count = {count1}, N = {nobs1}, Proportion = {prop1:.4f}")
    print(f"Sample 2: Count = {count2}, N = {nobs2}, Proportion = {prop2:.4f}")

    # Perform the Z-test for proportions
    # counts: array of successes [count1, count2]
    # nobs: array of total observations [nobs1, nobs2]
    z_statistic, p_value = proportions_ztest(count=[count1, count2], nobs=[nobs1, nobs2], alternative=alternative)

    print(f"\nZ-statistic: {z_statistic:.4f}")
    print(f"P-value ({alternative}): {p_value:.4f}")

    # Interpret the results
    interpretation = ""
    if p_value < alpha:
        interpretation = f"Since the P-value ({p_value:.4f}) is less than the significance level ({alpha}), we reject the null hypothesis."
        if alternative == 'two-sided':
            interpretation += "\nConclusion: There is a statistically significant difference between the two population proportions."
        elif alternative == 'smaller':
            interpretation += "\nConclusion: The proportion of the first group is significantly smaller than the second group."
        elif alternative == 'larger':
            interpretation += "\nConclusion: The proportion of the first group is significantly larger than the second group."
    else:
        interpretation = f"Since the P-value ({p_value:.4f}) is greater than the significance level ({alpha}), we fail to reject the null hypothesis."
        if alternative == 'two-sided':
            interpretation += "\nConclusion: There is no statistically significant difference between the two population proportions."
        elif alternative == 'smaller':
            interpretation += "\nConclusion: The proportion of the first group is not significantly smaller than the second group."
        elif alternative == 'larger':
            interpretation += "\nConclusion: The proportion of the first group is not significantly larger than the second group."

    return z_statistic, p_value, interpretation

# Example Usage 1: Two-sided test (e.g., comparing conversion rates of two website designs)
print("--- Example 1: Comparing Website Conversion Rates (Two-sided) ---")
conversions_A = 120 # Successes for Design A
visitors_A = 1000 # Total for Design A
conversions_B = 150 # Successes for Design B
visitors_B = 1100 # Total for Design B

z_stat, p_val, interp = z_test_proportions(conversions_A, visitors_A, conversions_B, visitors_B, alpha=0.05, alternative='two-sided')
print(f"\n{interp}")

# Example Usage 2: One-sided test (e.g., testing if a new drug is better than old drug)
print("\n--- Example 2: Comparing Drug Success Rates (One-sided, 'larger') ---")
successes_new_drug = 80
patients_new_drug = 100
successes_old_drug = 70
patients_old_drug = 100

# Testing if New Drug proportion is LARGER than Old Drug proportion
z_stat_one_sided, p_val_one_sided, interp_one_sided = z_test_proportions(
    successes_new_drug, patients_new_drug, successes_old_drug, patients_old_drug, alpha=0.05, alternative='larger'
)
print(f"\n{interp_one_sided}")

print("\n--- Significance of Z-test for Proportions ---")
print("This test is used when you want to compare the proportions (or percentages) of two independent groups.")
print("It's commonly applied in A/B testing, clinical trials, and surveys to see if observed differences in rates are statistically significant or due to random chance.")
print("Assumptions:")
print("- The samples are independent.")
print("- The observations within each sample are independent.")
print("- The sample sizes are large enough such that np >= 5 and n(1-p) >= 5 for both groups (where p is the pooled proportion). This ensures the sampling distribution of the proportion is approximately normal.")
```
